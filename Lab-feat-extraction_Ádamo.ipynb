```python

sns.set(style="whitegrid")
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
columns = ['customer_lifetime_value', 'income', 'monthly_premium_auto', 'total_claim_amount']

for i, col in enumerate(columns):
    sns.kdeplot(customer_df[col], shade=True, color=colors[i], ax=axes[i//2, i%2])
    axes[i//2, i%2].set_xlim(0, 2000)
    axes[i//2, i%2].set_xlabel(col, fontsize=12)
    axes[i//2, i%2].set_ylabel('Density', fontsize=12)
    axes[i//2, i%2].tick_params(axis='both', which='major', labelsize=10)

plt.tight_layout()
plt.show()
```

```python
for each in categorical:
    plt.figure(figsize=(12,8))
    sns.countplot(x = each, data=customer_df)
```

```python
from matplotlib.colors import Normalize
from matplotlib.cm import ScalarMappable


# Create a sequential color palette
cmap = plt.get_cmap('viridis')

# Create bar plots
for each in categorical:
    plt.figure(figsize=(12, 8))
    
    # Get unique categories and their counts
    categories = customer_df[each].unique()
    num_categories = len(categories)
    
    # Generate color map and normalize colors
    norm = Normalize(vmin=0, vmax=num_categories - 1)
    color_map = ScalarMappable(cmap=cmap, norm=norm)
    
    # Plot each category with a different color
    for i, category in enumerate(categories):
        color = color_map.to_rgba(i)
        category_counts = customer_df[customer_df[each] == category].groupby(each).size().reset_index(name='count')
        plt.bar(category_counts[each], category_counts['count'], color=color, label=category)
    
    # Set labels and title
    plt.xlabel(each, fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.title(f'Count of {each}', fontsize=14)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
  
    
    plt.show()

```

```python


# Define the variables
variables = ['customer_lifetime_value', 'income', 'monthly_premium_auto', 'total_claim_amount']
palette = sns.color_palette('husl', len(variables))
plt.figure(figsize=(12, 8))
for i, (var, color) in enumerate(zip(variables, palette), 1):
    plt.subplot(2, 2, i)
    sns.boxplot(y=var, data=customer_df, color=color)
    plt.xlabel(var, fontsize=12)
    plt.ylabel('Values', fontsize=12)
    plt.title(f'Boxplot of {var}', fontsize=14)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)

plt.tight_layout()
plt.show()


```
